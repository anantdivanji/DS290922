## Loading the Standard libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Load the dataset

```python
data = pd.read_csv('Health_Insurance.csv')
data.head()
```

```python
data.shape
```

## Seperate X and y

```python
X = data[['Age']]
y = data['Purchased']
```

## KFold Cross Validation from sklearn

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits = 10)
```

## Import Logistic Regression from sklearn

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

## Apply Logistic Regression on all set of train and test

```python
from sklearn.model_selection import cross_val_score
cross_val_score(model, X, y, cv = kf)
```

- Since number of splits mentioned is 10, hence the dataset will be split into 10 different train and test splits
- Using cross_val_score, we apply Logistic Regression on each train and test split.
- The output is the accuracy of Logistic Regression on each train and test splits

- From 10 accuracies, we observe that the 1 set of train and test split is providing the best accuracy which is 95%
