# Support Vector Regression

## Loading the standard libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Loading the data

```python
data = pd.read_csv('Position_Salaries.csv')
data.head()
```

# One Hot Encoding on Position

```python
data_ohe = pd.get_dummies(data['Position'], drop_first = True)
data_ohe
```

```python
data = pd.concat([data, data_ohe], axis = 1)
data.head()
```

```python
data.drop('Position', axis = 1, inplace = True)
data.head()
```

## Seperate X and y

```python
X = data.drop('Salary', axis = 1)
y = data['Salary']
```

## Divide the data into train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

## Fit the Support Vector Regressor on X_train, y_train

```python
from sklearn.svm import SVR
regressor = SVR()
```

```python
regressor.fit(X_train, y_train)
```

# Predictions on the X_test

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
X_test
```

## Model Evaluation 

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
