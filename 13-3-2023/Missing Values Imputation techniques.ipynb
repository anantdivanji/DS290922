## Missing Value Imputation using a constant

```python
data.fillna(48, inplace = True)
data
```

```python
data['Age'].fillna(48, inplace = True)
data
```

```python
data['Salary'].fillna(75000, inplace = True)
data
```

## Mean and Median Imputation

- Used only on Numerical columns

```python
data['Age'].mean()
```

```python
## Mean Imputation

data['Age'].fillna(data['Age'].mean(), inplace = True)
data
```

```python
round(data['Salary'].mean(), 2)
```

```python
mean_sal = round(data['Salary'].mean(), 2)
data['Salary'].fillna(mean_sal, inplace = True)
data
```

```python
## Median Imputation

data
```

```python
data['Age'].median()
```

```python
data['Age'].fillna(data['Age'].median(), inplace = True)
data
```

## It is better to go with median imputation as compared to mean imputation since Mean has an impact due to outliers

```python
data = pd.read_csv('Data_1.csv')
data
```

```python
data['Salary'].mean()
```

```python
data['Salary'].median()
```

## Mode Imputation

- Categorical data

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Embarked'].mode()
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace = True)
df.isnull().sum()
```

## Using sklearn to perform missing value imputation

```python
data
```

#### 1. Step I : import the library and invoke the strategy
#### 2. apply the strategy on the data to perform the imputation
#### Note: The output from this imputation technique will always be a numpy array

https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html

```python
from sklearn.impute import SimpleImputer
sim_imp = SimpleImputer(strategy = 'median')
```

```python
data = sim_imp.fit_transform(data[['Age', 'Salary']])
data
```

```python
data = pd.DataFrame(data, columns = ['Age', 'Salary'])
data
```
