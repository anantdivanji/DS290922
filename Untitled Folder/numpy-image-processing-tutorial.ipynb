I you are reading this notebook, then I am pretty sure you would have heard about OpenCV library for Images.
### Here is a fun fact: OpenCV treats image data as numpy's ndarray, so it is extremely useful to learn the processing in `NumPy` (ndarray)

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps

img = np.array(Image.open('sun.jpg'))
plt.figure(figsize=(8,8))
plt.imshow(img)
```

```python
print('# of dims: ',img.ndim)
print('img shape: ',img.shape)
print('dtype: ',img.dtype)
```

```python
# pixel value at [R, G, B]
print(img[20, 20])

# min pixel value at channel B
print(img[:, :, 2].min())
```

# 2. Rotate image

```python
img = np.array(Image.open('sun.jpg'))
img_grey = img.sum(2) / (255*3) # summing over axis=2 (channel axis) to get grey scaled image
img_grey
```

# 3. Negative of an image

```python
fig = plt.figure(figsize=(10, 10))
img_grey = 255*3 - img_grey        # 255 * 3 because we added along channel axis previously
fig.add_subplot(1, 2, 1)
plt.imshow(img_grey)
plt.title('Negative of Grey image')

img = 255 - img
fig.add_subplot(1, 2, 2)
plt.imshow(img)
plt.title('Negative of RGB image')
```

# 4. Padding black spaces

```python
img = np.array(Image.open('sun.jpg'))
img_grey = img.sum(2) / (255*3)
img0 = img_grey.copy()
img0 = np.pad(img0, ((100,100),(100,100)), mode='constant')
plt.imshow(img0)
```

# 5. Visualizing RGB channels

```python
img = np.array(Image.open('sun.jpg'))

img_R, img_G, img_B = img.copy(), img.copy(), img.copy()

img_R[:, :, (1, 2)] = 0
img_G[:, :, (0, 2)] = 0
img_B[:, :, (0, 1)] = 0



img_rgb = np.concatenate((img_R,img_G,img_B), axis=1)
plt.figure(figsize=(15, 15))
plt.imshow(img_rgb)
```

# 6. Color Reduction:


```python
img = np.array(Image.open('sun.jpg'))

# Making Pixel values discrete by first division by // which gives int and then multiply by same factor

img_0 = (img // 64) * 64    
img_1 = (img // 128) * 128

img_all = np.concatenate((img, img_0, img_1), axis=1)

plt.figure(figsize=(15, 15))
plt.imshow(img_all)
```

# 7. Trim Image by array slicing

```python
img = np.array(Image.open('sun.jpg'))

fig = plt.figure(figsize=(10, 10))

fig.add_subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original')

img0 = img[128:-128, 128:-128, :]

fig.add_subplot(1, 2, 2)
plt.imshow(img0)
plt.title('Trimmed')

```

# 14. Flip image


```python
img0 = img.copy()

for i in range(img0.shape[0] // 2):
    c = img0[i, :, :].copy()
    img0[i, :, :] = img0[img0.shape[0] - i - 1, :, :]
    img0[img0.shape[0] - i - 1, :, :] = c
        
plt.imshow(img0)
```

```python
## Or you can use :
plt.imshow(np.flipud(img)) #: [up/down]flips over x-axis

#### np.fliplr() : [left/right]flips over y_axis
```
