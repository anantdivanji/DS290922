## Loading the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Loading the dataset

```python
data = pd.read_csv('Social_Network_Ads.csv')
data.head(10)
```

```python
data.shape
```

## Checking for Missing Values

```python
data.isnull().sum()
```

## Divide the data into X(All independent variables) and y(target or dependent variable)

```python
X = data.drop('Purchased', axis = 1)
y = data['Purchased']
```

## Divide the data into train data and test data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
X_train.shape, X_test.shape
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

ss.fit_transform(X_train)
```

```python
ss.fit_transform(X_test)
```

## Apply the classification algorithm on the train data

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 9, metric = 'minkowski', p = 2)
knn.fit(X_train, y_train)
```

## Perform prediction on the test data

```python
y_pred = knn.predict(X_test)
y_pred
```

## Comparing the prediction and the test set results

```python
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

## Creating a confusion matrix and Checking accuracy of model

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_pred, y_test)
cm
```

## Calculate Precision, Recall, F1 score, error rate and Accuray

- True Positive = 20
- True Negative = 64
- False positive = 4
- False Negative = 12

```python
## Precision = TP / (TP + FP)

precision = 20 / (20 + 4)
precision
```

```python
recall = 20 / (20 + 12)
recall
```

```python
accuracy = (20 + 64) / (20 + 64 + 4 + 12)
accuracy
```

```python
error_rate = 1 - accuracy
error_rate
```

```python
error_rate = (4 + 12) / (4+12+20+64)
error_rate
```

```python
f1_score = (2 * precision * recall) / (precision + recall)
f1_score
```

## Accuracy calculation using sklearn.metrics

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```

## Precision, Recall, f1 score and accuracy using sklearn.metrics

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred, y_test))
```
