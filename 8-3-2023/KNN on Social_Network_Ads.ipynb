## Importing the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Load the dataset

```python
data = pd.read_csv('Social_Network_Ads.csv')
data.head()
```

## Check for missing values

```python
data.isnull().sum()
```

## Divide the data into X(Independent) and y(Dependent)

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
X.shape, y.shape
```

## Divide the data into train_test_split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
X_train.shape, X_test.shape
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

ss.fit_transform(X_train)
```

```python
ss.fit_transform(X_test)
```

## Apply the KNN Classifier Algorithm on train dataset

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)
```

## Check the predictions on the test dataset

```python
y_pred = knn.predict(X_test)
y_pred
```

## Comparing the prediction and test set results

```python
y_pred = knn.predict(X_test)
y_test = y_test.values
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Creating a Confusion Matrix and Checking

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_pred, y_test)
print(cm)
```

## Calculating Precision, Recall, F1 score and Accuracy

- True Negative: 69
- False Negative: 11
- False Positive : 10
- True positive: 30

```python
Precision = 30 / (30 + 10)
Precision
```

```python
recall = 30 / (30 + 11)
recall
```

```python
f1_score = (2* Precision * recall) / (Precision + recall)
f1_score
```

```python
Accuracy = (69 + 30) / (69 + 11 + 10 + 30)
Accuracy
```

## Checking Accuracy Score

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

## Printing the classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
